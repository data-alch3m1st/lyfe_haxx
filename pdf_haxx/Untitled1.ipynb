```python
import os
from PyPDF2 import PdfMerger  # Previously called PdfFileMerger

# List of PDF file paths you want to merge
pdf_files = [
    'file1.pdf'
    , 'file2.pdf'
    , 'file3.pdf'
]

# Optional: Sort by filename (if needed)
pdf_files.sort()

# Initialize merger
merger = PdfMerger()

# Loop through and append each PDF
for pdf in pdf_files:
    merger.append(pdf)
    print(f"Added: {pdf}")

# Generate output filename (optional logic)
output_name = "merged_output.pdf"

# Write to final output file
merger.write(output_name)
merger.close()

print(f"\nMerged PDF saved as: {output_name}")
```
