### PDF Compressor ###

`*This version is for scanned pdf's that are heavy in size...`

***and it doesnt cost 9.99/month ;)***

```python
# v0.0.2

import os
import tempfile
import pdfplumber
from PIL import Image

input_pdf = "../input/sample.pdf"          # <-- your original 7.2 MB file
output_pdf = "../output/compressed.pdf"    # <-- new smaller file

# Tuning knobs
DPI = 150          # adjust 120–200 depending on how sharp you want it
JPEG_QUALITY = 65  # adjust 50–80 (lower = smaller file)

temp_dir = tempfile.mkdtemp()
image_paths = []

# Render and compress each page as JPEG
with pdfplumber.open(input_pdf) as pdf:
    for i, page in enumerate(pdf.pages):
        print(f"Processing page {i+1}/{len(pdf.pages)}...")
        # Render the page as an image
        page_image = page.to_image(resolution=DPI).original
        page_image = page_image.convert("RGB")  # ensure compatibility with PDF
        
        img_path = os.path.join(temp_dir, f"page_{i:04d}.jpg")
        page_image.save(img_path, "JPEG", quality=JPEG_QUALITY, optimize=True)
        image_paths.append(img_path)

# Reassemble images into one PDF
print("Combining pages into final PDF...")
images = [Image.open(p) for p in image_paths]
first_page = images[0]
if len(images) > 1:
    first_page.save(output_pdf, save_all=True, append_images=images[1:])
else:
    first_page.save(output_pdf)

# Check results
orig_size = os.path.getsize(input_pdf) / 1024 / 1024
new_size = os.path.getsize(output_pdf) / 1024 / 1024
print(f"Done! Original: {orig_size:.2f} MB → New: {new_size:.2f} MB")
print(f"Output saved as: {output_pdf}")
```
